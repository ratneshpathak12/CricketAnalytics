```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#to display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)
```

```python
df = pd.read_csv('ODI_data.csv')
```

```python
df.head()
```

```python
len(df), len(df.columns)
```

```python
# Runs per innings
# SR
# 100's
# 50's
# Team contribution
```

```python
df['Innings Runs Scored Num'].unique()
```

```python
df = df[df['Innings Runs Scored Num'] != '-']
```

```python
df['Innings Runs Scored Num'].unique()
```

```python
df = df.dropna(subset = ['Innings Runs Scored Num'])
```

```python
df['Innings Runs Scored Num'].unique()
```

```python
df.head()
```

```python
#Convert to date time
df['Innings Date'] = pd.to_datetime(df['Innings Date'])
```

```python
df['year'] = df['Innings Date'].dt.year
```

```python
df.tail()
```

```python
df['Innings Runs Scored Num'] = df['Innings Runs Scored Num'].astype('int')
```

```python
df['Innings Balls Faced'] = df['Innings Balls Faced'].astype('int')
```

```python
df['Innings Not Out Flag'] = df['Innings Not Out Flag'].astype('int')
```

```python
# Sachin 1994 - 2005
# Virat 2009 - 2020
```

```python
sachin_df = df[(df.year >= 1994) & (df.year <= 2004)]
```

```python
virat_df = df[(df.year >= 2009) & (df.year <= 2019)]
```

```python
sachin_df.head()
```

```python
virat_df.head()
```

```python
# Criteria to Identify the Best Player
# Runs per innings = Total Runs scored/Total Innings played
# SR = 100*(Total Runs Scored/Total Balls played)
# 100's = sum(100's)
# 50's = sum(50's)
# No. of 4's
# No. of 6's
# Team contribution = Player Runs/Team Runs
```

```python
sdf = sachin_df[sachin_df['Innings Player'] == 'SR Tendulkar']
```

```python
sdf.head()
```

```python
sum(sdf['Innings Runs Scored Num'])
```

```python
vdf = virat_df[virat_df['Innings Player'] == 'V Kohli']
```

```python
vdf.head(20)
```

```python
sum(vdf['Innings Runs Scored Num'])
```

```python
# Number of Innings played by both
Innings_played_by_Sachin = "Innings played by Sachin=", len (sdf)
Innings_played_by_Virat = "Innings played by Virat=", len (vdf)
print(Innings_played_by_Sachin)
print(Innings_played_by_Virat)
```

```python
#Runs scored per Innings by both
Runs_scored_by_Sachin_Per_Innings = "Runs Scored by Sachin per Innings = ", sum(sdf['Innings Runs Scored Num'])/len(sdf)
Runs_scored_by_Virat_Per_Innings = "Runs Scored by Virat per Innings = ", sum(vdf['Innings Runs Scored Num'])/len(vdf)
print(Runs_scored_by_Sachin_Per_Innings)
print(Runs_scored_by_Virat_Per_Innings)
```

```python
#Strike Rate of both
Strike_Rate_of_Sachin = "Strike rate of Sachin =", 100*sum(sdf['Innings Runs Scored Num'])/sum(sdf['Innings Balls Faced'])
Strike_Rate_of_Virat = "Strike rate of Virat =", 100*sum(vdf['Innings Runs Scored Num'])/sum(vdf['Innings Balls Faced'])
print(Strike_Rate_of_Sachin)
print(Strike_Rate_of_Virat)
```

```python
# No. of 100's scored by both
Hundreds_by_Sachin = "Hundreds Scored by Sachin =", sum(sdf["100's"])
Hundreds_by_Virat = "Hundreds Scored by Virat =", sum(vdf["100's"])
print(Hundreds_by_Sachin)
print(Hundreds_by_Virat)
```

```python
# No. of 50's scored by both
Fifties_by_Sachin = "Fifties Scored by Sachin =", sum(sdf["50's"])
Fifties_by_Virat = "Fifties Scored by Virat =", sum(vdf["50's"])
print(Fifties_by_Sachin)
print(Fifties_by_Virat)
```

```python
sdf['Innings Boundary Sixes'] = sdf['Innings Boundary Sixes'].astype('int')
vdf['Innings Boundary Sixes'] = vdf['Innings Boundary Sixes'].astype('int')
sdf['Total Sixes'] = sdf['Innings Boundary Sixes'].astype('int')
vdf['Total Sixes'] = vdf['Innings Boundary Sixes'].astype('int')
```

```python
# No. of 6's scored by both
Number_of_Sixes_Scored_by_Sachin = "Sixes scored by Sachin from 1994 to 2004 =", sum(sdf['Total Sixes'])
print(Number_of_Sixes_Scored_by_Sachin)
Number_of_Sixes_Scored_by_Virat = "Sixes scored by Virat from 2009 to 2019 =", sum(vdf['Total Sixes'])
print(Number_of_Sixes_Scored_by_Virat)
```

```python
# No. of 4's scored by both
Number_of_Fours_Scored_by_Sachin = "Fours scored by Sachin from 1994 to 2004 =", sum(sdf['Total Fours']/4)
print(Number_of_Fours_Scored_by_Sachin)
Number_of_Fours_Scored_by_Virat = "Fours scored by Virat from 2009 to 2019 =", sum(vdf['Total Fours']/4)
print(Number_of_Fours_Scored_by_Virat)
```

```python
# No. of 6's scored per balled faced by both
Sixes_per_ball_faced_by_Sachin = "Sixes_per_ball_faced_by_Sachin =", (sum(sdf['Innings Balls Faced']))/sum(sdf['Total Sixes'])
print(Sixes_per_ball_faced_by_Sachin)
Sixes_per_ball_faced_by_Virat = "Sixes_per_ball_faced_by_Virat =", (sum(vdf['Innings Balls Faced']))/sum(vdf['Total Sixes'])
print(Sixes_per_ball_faced_by_Virat)
```

```python
# No. of 4's scored per balled faced by both
Fours_per_ball_faced_by_Sachin = "Fours_per_ball_faced_by_Sachin =", (sum(sdf['Innings Balls Faced']))/(sum(sdf['Total Fours'])/4)
print(Fours_per_ball_faced_by_Sachin)
Fours_per_ball_faced_by_Virat = "Fours_per_ball_faced_by_Virat =", (sum(vdf['Innings Balls Faced']))/(sum(vdf['Total Fours'])/4)
print(Fours_per_ball_faced_by_Virat)
```

```python
# Number of 6's scored by all Indian Players during Sachin's time 1994 to 2004
Percentage_of_Sixes_by_Sachin_as_compared_to_other_Indian_Players = "Percentage of sixes scored by Sachin as compared to other Indian players from 1994 to 2004 =", 100*(sum(sdf['Total Sixes'])/sum(sachin_df[sachin_df.Country == 'India']['Total Sixes']))
print(Percentage_of_Sixes_by_Sachin_as_compared_to_other_Indian_Players)
# Number of 6's scored by all Indian Players during Virat's time 2009 to 2019
Percentage_of_Sixes_by_Virat_as_compared_to_other_Indian_Players = "Percentage of sixes scored by Virat as compared to other Indian players from 2009 to 2019 =", 100*(sum(vdf['Total Sixes'])/sum(virat_df[virat_df.Country == 'India']['Total Sixes']))
print(Percentage_of_Sixes_by_Virat_as_compared_to_other_Indian_Players)
```

```python
# Number of 4's scored by all Indian Players during Sachin's time 1994 to 2004
Percentage_of_Fours_by_Sachin_as_compared_to_other_Indian_Players = "Percentage of Fours scored by Sachin as compared to other Indian players from 1994 to 2004 =", 100*(sum(sdf['Total Fours'])/sum(sachin_df[sachin_df.Country == 'India']['Total Fours']))
print(Percentage_of_Fours_by_Sachin_as_compared_to_other_Indian_Players)
# Number of 4's scored by all Indian Players during Virat's time 2009 to 2019
Percentage_of_Fours_by_Virat_as_compared_to_other_Indian_Players = "Percentage of Fours scored by Virat as compared to other Indian players from 2009 to 2019 =", 100*(sum(vdf['Total Fours'])/sum(virat_df[virat_df.Country == 'India']['Total Fours']))
print(Percentage_of_Fours_by_Virat_as_compared_to_other_Indian_Players)
```

```python
# Contribution of Sachin & Virat to Team Total

Contribution_of_Sachin_in_Team_Total = "Percentage of Runs scored by Sachin in India team from 1994 to 2004", 100*sum(sdf['Innings Runs Scored Num'])/sum(sachin_df[sachin_df.Country == 'India']['Innings Runs Scored Num'])
Contribution_of_Virat_in_Team_Total = "Percentage of Runs scored by Virat in India team from 2009 to 2019", 100*sum(vdf['Innings Runs Scored Num'])/sum(virat_df[virat_df.Country == 'India']['Innings Runs Scored Num'])
print(Contribution_of_Sachin_in_Team_Total)
print(Contribution_of_Virat_in_Team_Total)
```

```python
# Top Runs Scorer in Sachin's Generation from 1994 to 2004
sachin_df.groupby(['Innings Player'])['Innings Runs Scored Num'].sum().sort_values(ascending = False).head(10)
```

```python
# Visual Representation of Sachin's performance
sachin_df.groupby(['Innings Player'])['Innings Runs Scored Num'].sum().sort_values(ascending = False).head(10).plot(kind = 'barh', figsize = (10,6), title = 'Top Run scorers from 1994 to 2004')
plt.xlim(0,12000)
plt.xlabel('Number of Runs Scored', color = 'Red', fontsize='17', horizontalalignment='center')
plt.ylabel('Name of Players', color = 'Red', fontsize='17', horizontalalignment='center')
plt.show()
```

```python
# Top Runs Scorer in Virat's Generation from 2009 to 2019
virat_df.groupby(['Innings Player'])['Innings Runs Scored Num'].sum().sort_values(ascending = False).head(10)
```

```python
# Visual Representation of Virat's performance
virat_df.groupby(['Innings Player'])['Innings Runs Scored Num'].sum().sort_values(ascending = False).head(10).plot(kind = 'barh', figsize = (10,6), title = 'Top Run scorers from 2009 to 2019', xlabel ="Name of Players", ylabel = "Number of Runs scored")
plt.xlim(0,12000)
plt.xlabel('Number of Runs Scored', color = 'Red', fontsize='17', horizontalalignment='center')
plt.ylabel('Name of Players',  color = 'Red', fontsize='17', horizontalalignment='center')
plt.show()
```

```python
# Runs scored by Sachin Year on Year
sdf.groupby(['year'])['Innings Runs Scored Num'].sum().plot(kind = 'bar', title = "Sachin's performance over the years")
plt.xlabel('Years', color = 'Red', fontsize='17', horizontalalignment='center')
plt.ylabel('Number of Runs scored', color = 'Red', fontsize='17', horizontalalignment='center')
plt.axhline(y=1000)
plt.show()
```

```python
# Runs scored by Virat Year on Year
vdf.groupby(['year'])['Innings Runs Scored Num'].sum().plot(kind = 'bar', title = "Virat's performance over the years")
plt.xlabel('Years', color = 'Red', fontsize='17', horizontalalignment='center')
plt.ylabel('Number of Runs Scored', color = 'Red', fontsize='17', horizontalalignment='center')
plt.axhline(y=1000)
plt.show()
```
